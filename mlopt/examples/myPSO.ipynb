# RASTRIGIN

```python
def rastrigin(X, *args):
    return sum(list(map(lambda x: x**2 - 10*np.cos(2*np.pi*x) + 10, list(X))))
```

```python
qt_iter = 1000
xticks = np.arange(1, qt_iter+1,1)
dimentions = 30
```

## Global - Inercia Fixo

```python
myPSO_Ras_1 = PSO(number_of_particles=30, dim=dimentions, minx=-5, maxx=12, seed=1, function=rastrigin, functionArgs=[])
myPSO_Ras_1 = myPSO_Ras_1.Solver(max_epochs=qt_iter, plot_at_every=100, w=0.8, c1=2.05, c2=2.05, topology='G')
print("\nPSO completed\n")
ras_hist_fit_t1 = myPSO_Ras_1.historic_best_error
# fig= plt.figure(figsize=(10,6), dpi=300)
# plt.plot(xticks, ras_hist_fit_t1, linewidth=2)
# plt.xlabel("Iteração", fontsize=14)
# plt.ylabel("Valor objetivo", fontsize=14)
# plt.grid()
```

# ROSENBROCK

```python
def rosenbrock(X, *args):
    result = 0
    for i in range(len(X)-1):
        result += 100*(X[i+1]-X[i]**2)**2 + (X[i]-1)**2
    return result
```

## Global - Inercia Fixo

```python
myPSO_Ros_1 = PSO(number_of_particles=30, dim=dimentions, minx=-5, maxx=10, seed=1, function=rosenbrock, functionArgs=[])
myPSO_Ros_1 = myPSO_Ros_1.Solver(max_epochs=qt_iter, plot_at_every=100, w=0.8, c1=2.05, c2=2.05, topology='G')
print("\nPSO completed\n")
ros_hist_fit_t1 = myPSO_Ros_1.historic_best_error
# fig= plt.figure(figsize=(10,6), dpi=300)
# plt.plot(xticks, ros_hist_fit_t1, linewidth=2)
# plt.xlabel("Iteração", fontsize=14)
# plt.ylabel("Valor objetivo", fontsize=14)
# plt.grid()
```
